```python
import pandas as pd
import numpy as np
import Range_Model as rp # наша созданная модель
```

```python
#Импортируем Датасеты тестовой и тренировочной выборки
df_test = pd.read_csv('test.csv')
df_train = pd.read_csv('df_hack_fineal_fillna.csv')

df_train['MEAS_DT'] = pd.to_datetime(df_train['MEAS_DT'], unit= 's')
df_train

```

```python
# Получаем значения даты и времени из тестовый выборки, 
# чтобы в соответствии с ними получить признаки из тренировочной выборки для их использования при тестировании нашей модели


feats_name = ['MEAS_DT', 
              'Ni_1.1C', 'Ni_1.2C',     # Признаки, по которым будут строиться интервалы 
              'Cu_1.1C', 'Cu_1.2C',
              'Cu_2.1T', 'Cu_2.2T',
              'Cu_3.1T', 'Cu_3.2T',
              'Ni_4.1T', 'Ni_4.2T',
              'Ni_4.1C', 'Ni_4.2C',
              'Ni_5.1T', 'Ni_5.2T',
              'Ni_5.1C', 'Ni_5.2C',
              'Ni_6.1T', 'Ni_6.2T',
              'Ni_6.1C', 'Ni_6.2C']

# df_train['MEAS_DT'].isin(test_datetime).sum()

feats = df_train[df_train['MEAS_DT'].isin(df_test['MEAS_DT'])][feats_name]

# df_test = df_test.loc[df_test['MEAS_DT'].isin(feats['MEAS_DT'])]

# # print(df_train['Ni_1.1C'][df_train['MEAS_DT'].index (df_test['MEAS_DT'][0])])
feats
# feats_time = feats.MEAS_DT
# feats_time
```

```python
import math
dict_feats = {'Ni_1.1C': (0.1,	3, 0.2,	0.2),
              'Ni_1.2C': (0.1,	3, 0.2,	0.2),    
              'Cu_1.1C': (0.1,	5, 0.1,	0.2), 
              'Cu_1.2C': (0.1,	5, 0.1,	0.2),
              'Cu_2.1T': (0.01,	1, 0.2,	0.2),
              'Cu_2.2T': (0.01,	1, 0.2,	0.2),
              'Cu_3.1T': (0.05,	1, 0.3,	0.1), 
              'Cu_3.2T': (0.05,	1, 0.3,	0.1),
              'Ni_4.1T': (0.01,	2, 0.1,	0.2),
              'Ni_4.2T': (0.01,	2, 0.1,	0.2),
              'Ni_4.1C': (0.05,	4, 0.15,0.2),
              'Ni_4.2C': (0.05,	4, 0.15,0.2),
              'Ni_5.1T': (0.01,	2, 0.1,	0.2),
              'Ni_5.2T': (0.01,	2, 0.1,	0.2),
              'Ni_5.1C': (0.05,	4, 0.1, 0.2),
              'Ni_5.2C': (0.05,	4, 0.1, 0.2),
              'Ni_6.1T': (0.01,	1, 0.15, 0.01),
              'Ni_6.2T': (0.01,	1, 0.15, 0.01),
              'Ni_6.1C': (0.05,	3, 0.1, 0.01),
              'Ni_6.2C': (0.05, 3, 0.1, 0.01)}

for feat_name in feats_name[1:]:
    
    feat = df_train[feat_name]
    tracker = rp.AdaptiveIntervalTracker(
        granularity=dict_feats[feat_name][0], tolerance=dict_feats[feat_name][2], lock_threshold=8, max_buffer=dict_feats[feat_name][1], p=dict_feats[feat_name][3])

    data_stream = feat
    order = math.floor(math.log10(abs(dict_feats[feat_name][0])))
    current_interval_min = []
    current_interval_max = []
    for value in data_stream:
        current_interval = tracker.update_interval(value)
        current_interval_min.append(round(current_interval[0], abs(order)))
        current_interval_max.append(round(current_interval[1], abs(order)))

    df_train[feat_name + '_min'] = current_interval_min
    df_train[feat_name + '_max'] = current_interval_max
df_test["MEAS_DT"] = pd.to_datetime(df_test["MEAS_DT"])
df_train["MEAS_DT"] = pd.to_datetime(df_train["MEAS_DT"])
columns_to_merge = []
for i in feats_name[1:]:
    columns_to_merge.append(i + '_min')
    columns_to_merge.append(i + '_max')
merged_df = df_test.merge(df_train[["MEAS_DT"] + columns_to_merge], on="MEAS_DT", how="left", suffixes=("_old", ""))

# Обновление значений в df2 из merged_df
for col in columns_to_merge:
    merged_df[col] = merged_df[col].fillna(merged_df[f"{col}_old"])

# Удаление временных столбцов
merged_df = merged_df.drop(columns=[f"{col}_old" for col in columns_to_merge])
merged_df
```
